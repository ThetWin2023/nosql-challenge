```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

# Set up the query
query_London = {
    'LocalAuthorityName': {'$regex': 'London', '$options': 'i'},
    'RatingValue': {'$gte': 4}
}

# Search
result = establishments.find(query_London)
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that:
# 1. Matches establishments with a hygiene score of 0
query_hygiene = {'$match': {'scores.Hygiene' :0}}

# 2. Groups the matches by Local Authority
query_groupby = {'$group' : {'_id' : '$LocalAuthorityName', 'count' : {'$sum' : 1 }}}

# 3. Sorts the matches from highest to lowest
query_sortby =  {'$sort' : {'count':-1}}

# Limit to 10 results
query_limit = {'$limit':10}
```

```python
# Create the pipeline
pipeline = [query_hygiene , query_groupby, query_sortby, query_limit]
result_hyqiene_0_list = list(establishments.aggregate(pipeline))

# Print the first 10 results
pprint(result_hyqiene_0_list)
```

```python
# Convert the result to a Pandas DataFrame
result_hyqiene_0_df = pd.DataFrame(result_hyqiene_0_list)

# Display the number of rows in the DataFrame
len(result_hyqiene_0_df)
```
